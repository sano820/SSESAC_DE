5) 확인하고 수정!: 결과가 마음에 들지 않으면 수정 요청하기

```python
'''
문제 1. 비밀번호 강도 검사기
비밀번호를 입력받아 조건을 검사 후, 만족하는지 만족하지 않는지 판별 및 출력
항목을 모두 포함하는 프롬프트를 작성
● 역할: “파이썬 초보자를 가르치는 선생님”
● 입력: input()으로 비밀번호 입력
● 검사 기준: 길이 8자 이상, 숫자 포함(문자별 isdigit() 또는 any() 사용)
● 분기: 조건 모두 만족 → “안전한 비밀번호입니다”, 아니면 “약한 비밀번호입니다”
● 품질: if-else 사용, 단계별 주석, 의미 있는 변수명(password, length, has_number), 20줄 이내
● 동작 예시 3개 제시

프롬프트 {
    비밀번호를 입력받아 조건을 검사 후, 만족하는지 만족하지 않는지 판별 및 출력
    너는 파이썬 초보자를 가르치는 선생님이고 나는 파이썬 초보자 강의생이야. 
    아래 항목을 모두 포함하도록 python 코드를 작성해줘
    추가로 각 코드마다 주석을 달아서 설명 문도 작성해줘

    입력: input()으로 비밀번호 입력
    검사 기준: 길이 8자 이상, 숫자와 영어 둘 1글자 이상씩 포함, 같은 숫자나 문자가 3글자 연속 x
    분기: 조건 모두 만족 → “안전한 비밀번호입니다”, 아니면 “약한 비밀번호입니다”
    품질: if-else 사용, 단계별 주석, 의미 있는 변수명(password, length, has_number), 20줄 이내

    [입력]
    예시1 : "abc123"
    예시2 : "22222222"
    예시3 : "abcd1234"
    [출력]
    예시1 : "약한 비밀번호입니다."
    예시2 : "약한 비밀번호입니다."
    예시3 : "안전한 비밀번호입니다."
    
}
'''
# 비밀번호를 입력받음
password = input("비밀번호를 입력하세요: ")  # password 변수에 사용자 입력 저장

# 비밀번호 길이를 측정함
length = len(password)  # 길이를 의미하는 length 변수에 저장

# 숫자가 하나라도 포함되어 있는지 확인함
has_number = any(ch.isdigit() for ch in password)  # 숫자 포함 여부 판단

# 영문자가 하나라도 포함되어 있는지 확인함
has_alpha = any(ch.isalpha() for ch in password)  # 영문자 포함 여부 판단

# 같은 문자가 3번 연속되는지 검사하기 위한 초기값 설정
no_three_repeat = True  # 기본값은 연속 3글자 없음(True)

# 연속된 동일 문자가 3개 이상 있는지 반복문으로 검사
for i in range(length - 2):  # 인덱스 i, i+1, i+2를 비교하기 위해 length-2까지 반복
    if password[i] == password[i+1] == password[i+2]:  # 세 글자가 모두 동일하면
        no_three_repeat = False  # 연속 3문자 존재 표시
        break  # 더 이상 검사할 필요 없으므로 반복 종료

# 모든 조건(길이, 숫자 포함, 영문자 포함, 3연속 없음)이 만족되는지 판정
if length >= 8 and has_number and has_alpha and no_three_repeat:  # 모든 조건을 and로 연결
    print("안전한 비밀번호입니다.")  # 조건 만족 시 출력
else:
    print("약한 비밀번호입니다.")  # 하나라도 만족하지 않으면 출력
```

```python
'''
문제 2. 출석 체크 프로그램
학생 5명의 이름과 각자의 출석("출석"/"결석")을 입력 받아, 출석 인원과 출석률을 계산해 출력
항목을 모두 포함하는 프롬프트를 작성
● 역할: “초보자에게 코딩을 가르치는 선생님”
● 입력: 학생 5명 이름, 각자의 “출석/결석”
● 자료구조: 이름 리스트, 출결 리스트(각 5개)
● 계산: 출석 인원 수, 출석률 = (출석/5)*100 (소수점 1자리)
● 출력: 학생 5명 중 X명이 출석했습니다. 출석률: Y%
● 품질: 단계별 주석, 변수명(students, attendance), 30줄 이내, for로 5회 입력

프롬프트 {
    학생 5명의 이름과 각자의 출석("출석"/"결석")을 입력 받아, 출석 인원과 출석률을 계산해 출력

    ● 입력: 학생 5명 이름, 각자의 “출석/결석”
    ● 자료구조: 이름 리스트, 출결 리스트(각 5개)
    ● 계산: 출석 인원 수, 출석률 = (출석/5)*100 (소수점 1자리)
    ● 출력: 학생 5명 중 X명이 출석했습니다. 출석률: Y%
    ● 품질: 단계별 주석, 변수명(students, attendance), 30줄 이내

    각 코드마다 한국말로 친절히 주석을 달아줘
}
'''

# 학생 이름을 저장할 빈 리스트 생성
students = []

# 각 학생의 출석 여부를 저장할 빈 리스트 생성
attendance = []

# 총 학생 수를 상수로 설정 (5명)
total_students = 5

# 5명의 학생 이름과 출석 여부를 입력받기
for i in range(total_students):
    # i는 0부터 시작하므로 +1을 해줌 (1번 학생부터 표시)
    name = input(f"{i+1}번째 학생 이름을 입력하세요: ")
    # 이름을 리스트에 추가
    students.append(name)

    # 출석 여부 입력받기 ("출석" 또는 "결석")
    status = input(f"{name} 학생의 출석 여부를 입력하세요 (출석/결석): ")
    # 출석 여부를 리스트에 추가
    attendance.append(status)

# 출석 인원 수 계산 ("출석"으로 표시된 학생만 세기)
present_count = attendance.count("출석")

# 출석률 계산 (전체 대비 백분율)
attendance_rate = (present_count / total_students) * 100

# 결과 출력 (출석 인원과 출석률을 함께 표시)
# 소수점은 한 자리까지만 출력하도록 format() 사용
print(f"\n학생 {total_students}명 중 {present_count}명이 출석했습니다.")
print(f"출석률: {attendance_rate:.1f}%")

# 참고로, 리스트를 출력해서 확인하고 싶을 경우 (선택사항)
print("학생 이름 목록:", students)
print("출석 현황 목록:", attendance)
```

```python
'''
문제 3. 단어 빈도수 세기
문장 하나를 입력받아 단어를 분리하고, 각 단어의 등장 횟수를 세어 “사과: 3번”처럼 모든 단어와 빈도 출력
항목을 모두 포함하는 프롬프트를 작성
● 입력: 한 줄 문장
● 처리: split()으로 단어 리스트, 딕셔너리로 빈도 누적
● 출력: 단어: N번 형태로 전부 출력
● 품질: for + if-else(존재 시 +1, 없으면 1), in 사용, 25줄 이내, 줄마다 주석

프롬프트 {
    문장 하나를 입력받아 단어를 분리하고, 각 단어의 등장 횟수를 세어 “사과: 3번”처럼 모든 단어와 빈도 출력
    
    ● 입력: 한 줄 문장
    ● 처리: split()으로 단어 리스트, 딕셔너리로 빈도 누적
    ● 출력: 단어: N번 형태로 전부 출력
    ● 품질: for + if-else(존재 시 +1, 없으면 1), in 사용, 25줄 이내, 줄마다 주석
}
'''
# 문장 하나를 입력받기
sentence = input("문장을 입력하세요: ")

# 입력받은 문장을 공백 기준으로 단어로 나누기
words = sentence.split()

# 단어의 등장 횟수를 저장할 빈 딕셔너리 생성
word_count = {}

# 각 단어를 하나씩 꺼내서 빈도 세기
for word in words:
    # 만약 단어가 딕셔너리에 이미 있다면
    if word in word_count:
        # 기존 값(횟수)에 1을 더함
        word_count[word] += 1
    # 만약 단어가 처음 등장한 경우라면
    else:
        # 새 단어로 딕셔너리에 추가하고 값은 1로 설정
        word_count[word] = 1

# 모든 단어와 등장 횟수를 출력
# 딕셔너리의 key, value를 함께 꺼내서 출력
for word, count in word_count.items():
    print(f"{word}: {count}번")
```

```python
'''
문제 5. 파일 이름 정리 프로그램
파일 이름 리스트에서 .jpg 파일만 골라 알파벳순으로 정렬해 출력
항목을 모두 포함하는 프롬프트를 작성
● 입력 데이터: files = ["photo1.jpg","document.pdf","photo2.jpg","music.mp3","image.jpg"]
● 필터: .endswith(".jpg") 로 .jpg만 선별
● 정렬: 알파벳순
● 출력: 목록 형태
● 품질: for + if, 단계별 주석, 20줄 이내

프롬프트 {
    파일 이름 리스트에서 .jpg 파일만 골라 알파벳순으로 정렬해 출력

    입력 데이터: files = ["photo1.jpg","document.pdf","photo2.jpg","music.mp3","image.jpg"]
    필터: .endswith(".jpg") 로 .jpg만 선별
    정렬: 알파벳순
    출력: 목록 형태
    품질: for + if, 단계별 주석, 20줄 이내, 함수화
}
'''
def get_jpg_files(files):                 # 파일 리스트를 받아 .jpg 파일만 반환하는 함수 정의
    jpg_list = []                         # .jpg 파일을 담을 빈 리스트 생성
    for f in files:                       # 입력된 파일명 리스트를 하나씩 반복 검사
        if f.endswith(".jpg"):            # 파일명이 ".jpg"로 끝나는지 확인
            jpg_list.append(f)            # 조건에 맞으면 jpg_list에 추가
    jpg_list.sort()                       # 모은 jpg 파일들을 알파벳순으로 정렬
    return jpg_list                       # 정렬된 리스트를 반환

# 예시 입력 데이터
files = ["photo1.jpg", "document.pdf", "photo2.jpg", "music.mp3", "image.jpg"]  # 파일명 샘플

# 함수 호출하여 결과를 받고 출력
result = get_jpg_files(files)             # 함수 호출로 .jpg 파일 목록을 얻음
print("jpg 파일 목록:", result)           # 목록 형태로 결과 출력
```

```python
'''
문제 6. 간단한 할 일 목록(To-Do)
메뉴(1 추가, 2 목록, 3 완료 표시, 4 종료)를 반복해서 보여주고 사용자 선택에 따라 동작
항목을 모두 포함하는 프롬프트를 작성
● 메뉴: 1 추가, 2 목록, 3 완료 표시, 4 종료
● 동작: 리스트에 추가/출력, 인덱스로 완료 표시(앞에 "완료: ")
● 반복: while True + if-elif-else
● 품질: enumerate()로 번호, 기능별 주석, 40줄 이내

프롬프트 {
    할 일 목록 만들기를 할꺼야.
    메뉴(1 추가, 2 목록, 3 완료 표시, 4 종료)를 반복해서 보여주고 사용자 선택에 따라 동작

    ● 메뉴: 1 추가, 2 목록, 3 완료 표시, 4 종료
    ● 동작: 리스트에 추가/출력
    ● 반복: while True + if-elif-else
    ● 품질: enumerate()로 번호, 기능별 주석, 40줄 이내

    [입력]
    1
    운동하기
    2
    1
    공부하기
    3
    운동하기
    2
    4
    [출력]
    목록에 운동하기가 추가 되었습니다.
    목록 : 운동하기
    목록에 공부하기가 추가 되었습니다.
    목록 : 운동하기, 공부하기
    운동하기를 완료하였습니다.
    목록 : 공부하기
    To-Do 리스트가 종료되었습니다.

}
'''
# To-Do 리스트 관리 프로그램

# 할 일을 저장할 빈 리스트 생성
todo_list = []
# 사용자에게 메뉴 선택 안내 출력
print("\n메뉴를 선택하세요:")
print("1. 추가  2. 목록 보기  3. 완료 표시  4. 종료")

# 무한 반복으로 메뉴를 계속 보여줌
while True:
    # 메뉴 번호 입력받기
    choice = input("번호를 입력하세요: ")

    # 1번 선택 → 새로운 할 일 추가
    if choice == "1":
        task = input("추가할 일을 입력하세요: ")      # 추가할 일 입력
        todo_list.append(task)                        # 리스트에 추가
        print(f"목록에 {task}가 추가 되었습니다.")     # 추가 완료 메시지 출력

    # 2번 선택 → 현재 목록 출력
    elif choice == "2":
        if len(todo_list) == 0:                       # 목록이 비어 있을 경우
            print("현재 할 일이 없습니다.")
        else:
            print("현재 목록:")
            # enumerate로 번호와 함께 출력
            for idx, task in enumerate(todo_list, 1):
                print(f"{idx}. {task}")

    # 3번 선택 → 완료한 일 표시 (삭제)
    elif choice == "3":
        if len(todo_list) == 0:                       # 완료 표시할 항목이 없을 때
            print("완료할 일이 없습니다.")
        else:
            done = input("완료한 일을 입력하세요: ")
            if done in todo_list:                     # 입력한 일이 목록에 있으면 삭제
                todo_list.remove(done)
                print(f"{done}를 완료하였습니다.")
            else:
                print(f"{done}는 목록에 없습니다.")

    # 4번 선택 → 프로그램 종료
    elif choice == "4":
        print("To-Do 리스트가 종료되었습니다.")
        break                                         # 반복문 종료

    # 잘못된 입력 처리
    else:
        print("잘못된 번호입니다. 1~4 사이에서 선택하세요.")
```

```python
'''
문제 7. 숫자 야구 게임
0~9에서 서로 다른 수 3개를 뽑아 사용자가 맞추는 숫자 야구 게임 구현
항목을 모두 포함하는 프롬프트를 작성
● 목표: 서로 다른 0~9 숫자 3개 맞추기
● 컴퓨터 숫자: 랜덤 숫자
● 입력: "123" 같은 문자열 → 각 자리 int 리스트로 변환
● 판정: Strike(숫자+위치), Ball(숫자만), Out(없음)
● 반복: 맞출 때까지 while
● 품질: 단계별 자세한 주석, 40줄 이내

프롬프트 {
    숫자 야구 게임 : 0~9에서 서로 다른 수 3개를 뽑아 사용자가 맞추는 숫자 야구 게임 구현

    ● 목표: 서로 다른 0~9 숫자 3개 맞추기
    ● 컴퓨터 숫자: 랜덤 숫자
    ● 입력: "123" 같은 문자열 → 각 자리 int 리스트로 변환
    ● 판정: Strike(숫자+위치), Ball(숫자만), Out(없음)
    ● 반복: 맞출 때까지 while
    ● 품질: 단계별 자세한 주석, 40줄 이내
}
'''

import random  # 랜덤 숫자 생성을 위해 random 모듈을 불러옵니다.

# 컴퓨터가 뽑을 서로 다른 3개의 숫자를 생성합니다.
secret = random.sample(range(10), 3)  # 0~9 중 서로 다른 3개를 랜덤 선택하여 리스트로 만듭니다.

# 시도 횟수를 세기 위한 변수 초기화
attempts = 0

print("숫자 야구 게임을 시작합니다! 0~9 사이 서로 다른 숫자 3개를 맞춰보세요.")
print("입력 형식 예: 123 (각 자리 숫자는 중복되면 안 됩니다.)\n")

# 사용자가 3스트라이크(정답)를 맞출 때까지 반복합니다.
while True:
    attempts += 1  # 시도 횟수 증가

    # 사용자의 입력을 받습니다.
    user_input = input(f"{attempts}번째 시도 - 숫자 3개를 입력하세요: ")

    # 입력 검증: 길이 3, 모두 숫자, 각 자리 중복 없음
    if len(user_input) != 3 or not user_input.isdigit():
        print("잘못된 입력입니다. 반드시 숫자 3자리(예: 123)로 입력하세요.\n")
        continue

    # 문자열을 정수 리스트로 변환 (예: "123" -> [1,2,3])
    guess = [int(ch) for ch in user_input]

    # 입력 숫자들에 중복이 있는지 검사
    if len(set(guess)) != 3:
        print("중복된 숫자가 있습니다. 서로 다른 숫자 3개를 입력해주세요.\n")
        continue

    # 스트라이크와 볼 카운트 초기화
    strike = 0
    ball = 0

    # 판정 로직: 각 자리 비교로 스트라이크 세기, 숫자만 존재하면 볼 세기
    for i in range(3):
        if guess[i] == secret[i]:
            strike += 1
        elif guess[i] in secret:
            ball += 1

    # 결과 출력: 스트라이크, 볼, 또는 아웃
    if strike == 3:
        print(f"{attempts}번째 시도: {strike}스트라이크! 정답입니다. 축하합니다!")
        break  # 정답이면 반복 종료
    elif strike == 0 and ball == 0:
        print(f"{attempts}번째 시도: 아웃!\n")
    else:
        print(f"{attempts}번째 시도: {strike}스트라이크, {ball}볼\n")
```
