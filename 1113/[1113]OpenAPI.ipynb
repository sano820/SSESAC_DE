# OpenAPI

- 누구나 사용할 수 있도록 공개된 api를 말하며, 웹 사이트가 가진 기능을 모두가 이용할 수 있도록 공개한 프로그래밍 인터페이스

## API(Application Programming Interface)

- 응용 프로그램끼리 서로 소통하기 위한 약속된 규칙
    - 하나의 프로그램이 다른 프로그램의 기능이나 데이터를 표준화된 방식으로 요청할 수 있도록 하는 통로

### API 구조

기본적인 구조
- 웹 기반 클라이언트-서버 간의 요청/응답 구조


API 구조 (1) : End-point
- 클라이언트가 서버에 요청을 보낼 수 있는 정확한 주소


API의 구조(2): Query parameter(=Query String)
- 요청시, 웹 서버에 추가로 제공하는 파라미터(필터링, 옵션)
    - Endpoint 주소 끝에 ?로 시작한다.
    - "파라미터이름 = 값" 형태로 전달하며, 여러개의 파라미터를 넣는 경우에는 파라미터마다 &로 구분
