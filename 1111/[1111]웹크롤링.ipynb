## 웹 크롤링(Web Crawling)

- 인터넷 상의 웹 페이지를 자동으로 탐색하고 데이터를 수집하는 기술
    - 공식적인 API가 없거나 제한적일 때 유용
    - 공개된 웹 페이지의 가격/평판/리뷰/리스트 등과 같은 반정형 데이터(HTML)를 신속히 모을 때 유용
- 웹 페이지를 가져와서 필요한 부분만 추출
    - 자동화
    - 대량 데이터 수집
    - 구조화된 데이터 추출

### 웹 크롤링: 웹 페이지에서 데이터 추출
- 웹 페이지는 HTML라는 마크업 언어로 이루어져 있다.
- 웹 페이지는 HTML, CSS, JavaScript의 조합으로 구성
    - HTML(Hyper Text Markup Language) : 웹 페이지의 구조, 뼈대/골격을 담당(텍스트, 이미지, 링크, 표, 리스트 등)
    - CSS: 웹 페이지의 디자인을 담당
    - JavaScript: 웹 페이지의 동적(Interaction)을 담당
- 원하는 주소의 웹페이지로 들어가 HTML을 가져오고, 그 안에서 원하는 데이터가 어디에 있는지 가져오는 작업(Parsing)

## HTML(Hypertext Markup Language)의 이해

### HTML 기본 구성 요소 : Tag

- 태그(Tag)는 HTML 문서를 구성하는 최소 단위의 블록

- 속성(Attribute)는 해당 요소에 추가적인 옵션을 사용할 때(글자색, 크기, 배경색 등)


### HTML의 다양한 태그(1): 시맨틱 태그(Semantic Tag)

의미와 구조가 담겨있는 태그(실질적 표현 x)


### HTML의 다양한 태그(2) : 콘텐츠 태그(Contents Tag)

실제 내용을 표현하는 태그로 사용자에게 보이는 데이터를 표현


### HTML의 다양한 태그(3) : 그 외의 비시맨틱 태그(Non-Sematic Tag)


- 의미 없이 스타일/레이아웃을 위한 태그


### HTML의 계층 구조(Hierarchy Structure)

HTML은 부모(Parent)와 자식(Child) 태그로 이루어진 트리(Tree) 형태의 계층 구조
- 태그 안에 또 다른 태그가 들어갈 수 있고, 그 안에 또 다른 하위 요소가 계속 중첩된다.

## CSS(Cascading Style Sheets)의 이해

CSS의 적용 방식
1. 인라인 방식(태그의 속성에 직접 부여)
2. style 태그에 내부 스타일 지정
3. 외부 스타일 불러오기(별도의 .css 정의)

### CSS의 기본 구조

- 세모콜론: 각 속성 선언의 끝을 표시
- 중괄호: 선택자와 속성 블록을 구분


### CSS의 다양한 선택자(Selector) (1): 요소 선택자

- 어떤 요소를 꾸밀지를 정하는 "대상 지정 문법"


### CSS의 다양한 선택자(Selector) (2): 아이디 선택자

- 한 문서에서 고유한 이름으로, 특정 하나의 요소를 식별(#id 형태의 CSS 선택자)


### CSS의 다양한 선택자(Selector) (3): 클래스 선택자

- 공통된 그룹 이름으로, 여러 요소에 공통적으로 사용 가능(.clss 형태의 CSS 선택자)
- 크롤링에서 엄청 많이 사용됨.


### CSS의 다양한 선택자 (4) : 자식 선택자

- 부모의 직접적인 자식 요소에만 적용


### CSS의 다양한 선택자 (5): 가상 클래스 선택자


- 요소의 특정 상태나 조건에 따라 적용
- 종류가 많으니 나중에 찾아서 해보기
- 아래 이미지 >로 수정하기

## Beautiful Soup 활용하기

HTTP 통신
- 웹 브라우저와 웹 서버 사이에 데이터를 주고 받는데 사용되는 통신  
client -> 요청(Request) -> server  
cliend <- 응답(Response) <- server  


크롤릴에 필요한 요소 : Requests 라이브러리
- 접근할 웹 페이지의 데이터를 요청/응답 받기 위한 라이브러리

- Get 요청
    - 웹 서버에게 파라미터를 포함해 요청을 보내는 가장 쉬운 방법, URL에 파라미터 정보를 담아서 보내는 것
        - 사용자가 서버에게 웹페이지 보여 달라는 것이 요청(Requet)
        - 서버가 요청에 대한 대답과 HTML 문서를 주는 것을 응답(Response)

Beautiful Soup
- HTML 및 XML 등 정적 문서를 구문을 분석하기 위한 라이브러리
    - HTML 문서를 파이썬이 읽기 쉬운 구조로 변환하여 원하는 태그/텍스트/속성 등을 쉽게 구조적으로 추출(Parsing)하는 도구
    - !단, 정적 HTML에 대해서만 적용 가능
        - Selenium 활용하면 가능

라이브러리 설치 -> 라이브러리 불러오기 -> 웹페이지 요청 후, BeautifulSoup 객체 생성

```python
# pip install requests
# pip install bs4

import requests                 # 웹 페이지 요청을 위해
from bs4 import BeautifulSoup   # 가져온 문서에서 데이터 추출을 위해

url = "https://poke-match-type.site/"                        # 크롤링하고자 하는 페이지의 URL
res = requests.get(url)         # 해당 URL에 대해 GET 요청
bs = BeautifulSoup(res.text)    # 받은 응답 텍스트(HTML)을 BeautifulSoup 클래스의 인자로 전달 후, 객체 생성(bs)
print(bs)
```
