## 문제 4. 

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
    <style>
        .blue{
            color : blue;
        }
        .red{
            color: red;
        }
    </style>
</head>
<body>
    <p class="blue">이 글자는 파란색입니다.</p>
    <p class="red">이 글자는 빨간색입니다.</p>
</body>
</html>
```

## 종합 문제: 간단한 프로필 카드 만들기

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>내 프로필</title>
    <style>
        h1 {
            font-size: 30px;
            color: blue;
        }
        h2 {
            font-size: 20px;
            color: black;
        }
        .intro{
            border: 2px dotted gray;
            padding: 15px;
            background-color: lightgray;
        }
    </style>
</head>
<body>
    <h1>내 프로필</h1>
    <h2>박상선</h2>
    <div class = "intro">
        <p>안녕하세요. 박상선입니다.</p>
        <h3>관심사</h3>
        <ul>
            <li>코딩</li>
            <li>축구</li>
            <li>잠자기</li>
        </ul>
    </div>
</body>
</html>
```
