# 데이터 시각화

## 이론.

데이터 시각화의 목적과 필요성
- 궁극적으로 문제해결 또는 주장을 위함.
- 결과를 바탕으로 해석 후, 액션을 취하기 위함
    - 숫자만 봐서는 보이지 않은 흐름을 시각적으로 포척
    - 여러 집단, 시점, 변수 간 차이를 한눈에 비교 가능(패턴 파악)
    - 분석으로 얻은 인사이트를 통해 경영, 마케팅, 연구 등에 의사결정 지원.

### 다양한 데이터의 유형

- 데이터는 센서, 설문지 등 다양한 장소, 다양한 형태로 저장
- 데이터는 숫자로 표현되는지를 중점으로 구분.
    - 정량적 데이터
        - 연속형과 이산형
    - 정성적 데이터
        - 순서에 따라 순서형과 명목형으로 나뉨.
    - 텍스트와 날짜

**정량적 데이터(Quantitative Data)**
- 연속형 데이터(Continuous Data)
    - 연속적인 볌위 내에서 측정되는 데이터(셀 수 없음)
        - ex. 길이, 몸무게, 기온 등
- 이산형 데이터
    - 분리된 개별 값으로 측정되는 데이터(셀 수 있음)
        - ex. 사람 수, 과일 개수 등

**정성적 데이터(Qualitative Data)**
- 범주 순서형 데이터(Ordinal Data)
    - 범주 간의 상대적인 순서 또는 등급을 가지고 데이터
        - ex. 좋음/보통/나쁨, 5점/4점///1점
- 범주 비순서형 데이터(명목형)(Nomial Data)
    - 범주 또는 카테고리를 나타내는 데이터(순서 X)
        - ex. 남자/여자, 혈액형 A/B/O/AB

### 데이터 유형에 따른 시각화 그래프

## Matplotlib 라이브러리

가장 기본적인 저수준의 그래프를 그리는 파있너의 시각화 라이브러리

````python
# 설치하기
# !pip install matplotlib
# matplotlib 불러오기
import matplotlib.pyplot as plt

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
````

### 선 그래프 그리기

````python
year = [2018, 2019, 2020, 2021, 2022, 2023]
python = [5.8, 8.17, 9.31, 11.87, 15.63, 14.51]
C = [13.59, 14.08, 16.72, 14.32, 12.71, 14.73]
java = [15.78, 15.04, 16.73, 11.23, 10.82, 13.23]
JS = [3.49, 2.51, 2.38, 2.44, 2.12, 2.1]
````

````python
'''
선 그래프 그리기 : 연도에 따른 파이썬 점유율 변화

plot() 함수: 선 그래프 그리기
- x축 데이터와 y축 데이터를 리스트 형태로 전달

plt.show() 함수: 그래프 출력
'''

plt.plot(year, python)
plt.show()
````

````python
'''
제목(title)과 label 표시하기

title(): 그래프 제목 설정
xlabel(): x축 레이블 설정
ylabel(): y축 레이블 설정

fontsize 매개변수로 글자 크기 조절 가능
'''

plt.title("파이썬의 언어 점유율", fontsize=20)

plt.xlabel("연도")
plt.ylabel("점유율", fontsize=15)  

plt.plot(year, python)
plt.show()
````

````python
'''
여러 그래프를 함께 표시하기
'''
plt.plot(year, python)
plt.plot(year, C)
plt.plot(year, java)
plt.show()
````

````python
'''
범례(legend) 표시하기(1)

그래프가 그려지는 순서에 맞게끔 legend()의 인자로 전달
'''
plt.plot(year, python)
plt.plot(year, C)
plt.plot(year, java)

plt.legend(['파이썬', 'C', '자바'])
````

````python
'''
범례(legend) 표시하기(2)

그래프의 label 매개변수에 범례 이름 지정
단, legend() 함수는 반드시 그래프를 그린 후에 호출
'''
plt.plot(year, python, label='파이썬')
plt.plot(year, C, label='C')
plt.plot(year, java, label='자바')
plt.legend()
plt.show()
````

## 그림이 그려지는 구조

Figure
- 그림을 그리기 위한 Canvas의 역할(도화지)
  
Axes
- x축과 y축이 있는 그래프를 그리는 좌표 공간
- 실제 그림을 그리는 메서드들을 가짐.
    - 순서대로 호출되는 메서드들이 하나씩 도화지 위에 그려짐.
  
최종적으로 Figure 그리고 Axes에 쌓인 요소들을 한번에 렌더링 하는 것이 plt.show()

### 방법1. 여러 개의 Figure를 사용하기

plt.figure()를 호출할 때마다 새로운 도화지를 생성 (ID 부여 가능)

````python
fig1 = plt.figure(1, figsize=(4,3)) # ID가 1인 새로운 figure 객체를 생성하고 선택
# 앞으로 호출하는 함수는 fig1에 그려짐
plt.plot(year, python, label='파이썬') # 현재 선택된 fig1에 선 그래프르르 그림
plt.title("fig1") # fig1에 제목 설정

fig2 = plt.figure(2, figsize=(4,3)) # ID가 2인 새로운 figure 객체를 생성하고 선택
# 앞으로 호출하는 함수는 fig2에 그려짐
plt.plot(year, java, label='JAVA') # 현재 선택된 fig2에 선 그래프르르 그림
plt.title("fig2") # fig2에 제목 설정

plt.show()
````

````python
'''
plt.figure(id)로 특정 Figure 선택하기
- id에 해당하는 Figure가 존재하면 해당 Figure를 선택
'''
fig1 = plt.figure(1, figsize=(4,3))
plt.plot(year, python, label='파이썬') 
plt.title("fig1")

fig2 = plt.figure(2, figsize=(4,3))
plt.title("fig2")

plt.figure(1)  # ID가 1인 Figure 선택(없으면 새로 생성)
plt.plot(year, java, label='JAVA') # 현재 활성화된 figure에 선그래프 그림

plt.show()
````

### 방법2. 여러 개의 figure를 여러 공간으로 나누기:subplots

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,6))
# 1행 2열의 서브플롯 생성
# ax1: 첫 번째 서브플롯, ax2: 두 번째 서브플롯

ax1.plot(year, python)
ax2.plot(year, java)

ax1.set_title("파이썬 그래프")
ax2.set_title("자바 그래프")
plt.show()
````

subplots() 생성 여러 케이스
- 도화지 한 장에 하나의 공간
    - ```fig, ax = plt.subplots()```
- 도화지 한장에 1행 2열 (가로 2개)
    - ```fig, ax = plt.subplots(1,2, figsize=(10,4))```
    - 반환되는 ax 2개의 원소를 갖는 리스트 형태
        - ```ax[0].plt([1,2,3],[1,4,9])```
        - ```ax[1].plt([1,2,3],[9,4,1])```
- 도화지 한 장에 2행 2열(가로 2개, 세로 2개)
- 반환되는 ax는 2차원 리스트 형태 // ax[행, 열] 형태로 접근
    - ```fig, ax = plt.subplots(2,2, figsize=(8,6))``
        - ```ax[0,0].plot([1,2,3],[3,2,1])```
        - ```ax[0,1].bar([1,2,3],[3,2,1])```
        - ```ax[1,0].hist([1,2,3,2,1])```
        - ```ax[1,1].scatter([1,2,3],[3,2,1])```

````python
fig, ax = plt.subplots(2,2, figsize=(8,6))
ax[0,0].plot([1,2,3],[3,2,1])
ax[0,1].bar([1,2,3],[3,2,1])
ax[1,0].hist([1,2,3,2,1])
ax[1,1].scatter([1,2,3],[3,2,1])
````

### subplot 간의 겹치는 문제?

````python
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(year, python)
ax2.plot(year, java)

ax1.set_title('파이썬 점유율')
ax2.set_title('자바 점유율')
````

````python
'''
constrained_layout() 매개변수로 subplot 간의 겹침 문제 해결
'''
fig, (ax1, ax2) = plt.subplots(2,1, constrained_layout=True )
ax1.plot(year, python)
ax2.plot(year, java)

ax1.set_title('파이썬 점유율')
ax2.set_title('자바 점유율')
````

## 여러가지 옵션

### 눈금 조절하기

````python
'''
여러가지 옵셥(1) : 눈금 조절하기

눈금의 숫자를 리스트 형태로 전달
xticks(): x축 눈금 설정
yticks(): y축 눈금 설정

rotation 매개변수로 눈금 글자 회전 가능
'''
plt.plot(year, python, label='파이썬')

plt.xticks([2018,2019,2020,2021,2022,2023],rotation = 90)
plt.yticks(range(0,17), rotation=45)

plt.show()
````

## 다양한 그래프 그리기(1) : 선 그래프

선그래프

    - 시간, 순서, 변화 등에 따른 데이터의 추이와 패턴을 시각적으로 파악하기에 유용

````python
fig, (bx1, bx2) = plt.subplots(1,2)
sns.lineplot(x=year, y=python, ax=bx1)
bx2.plot(year, python)
````

````python
import pandas as pd
df = pd.DataFrame(pd.read_csv('data/seoul_park.csv'))

# 월별로 그룹화하여 평균값 계산
grouped_mon = df.groupby("월")
# 각 월별로 수치형 컬럼들의 평균을 계산
df_mon = grouped_mon.mean(numeric_only=True)
df_mon
````

````python
'''
seaborn으로 선그래프 그리기

- 직접 그룹화 없이 'data'에 데이터 프레임 전달
- x,y에는 사용할 컬럼 이름만 전달
'''
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

fig, (cx1,cx2) = plt.subplots(1,2, constrained_layout=True)
# ---------------------
# (1) matplotlib 선그래프
# --------------------
cx1.set_title("서울대공원 방문자 수 비교 (matplotlib)")
cx1.plot(df_mon['어른'], label='어른', marker='o', linestyle='--', color='b')
cx1.plot(df_mon['어린이'], label='어린이', marker='s', linestyle='-.', color='r')
cx1.plot(df_mon['청소년'], label='청소년', marker='^', linestyle=':', color='g')
cx1.set_xticks(range(1,13))
cx1.set_xlabel("월")
cx1.set_ylabel("방문자 수")
cx1.legend()
cx1.grid(True, axis = 'x')
# ---------------------
# (2) seaborn 선그래프
# --------------------
cx2.set_title("서울대공원 방문자 수 비교 (seaborn)")
sns.lineplot(data=df, x='월', y ='어른',label = '어른', ax = cx2)
# errorbar=None 옵션 전달시, 자동으로 나오는 오차 영역/막대 등 삭제
sns.lineplot(data=df, x='월', y ='어린이',label = '어린이', errorbar=None, ax = cx2)
sns.lineplot(data=df, x='월', y ='청소년',label = '청소년', errorbar=None, ax = cx2)
cx2.set_xticks(range(1,13))
cx2.set_xticklabels(['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월'], rotation=45)
plt.show()
````

## 다양한 그래프 그리기(2) : 막대 그래프

막대 그래프
- x축과 y축을 갖는 좌표 평면 상에 데이터를 막대로 표현한 그래프
    - 복수의 데이터를 비교하는데 유용한 그래프
    - 값의 크고 작음을 한 눈에 파악하기에 유용한 시각화 상태

````python
fig, (bx1, bx2) = plt.subplots(1,2)
sns.barplot(x=year, y=python, ax=bx1)
bx2.bar(year, python)
````

````python
import pandas as pd
df = pd.DataFrame(pd.read_csv('data/seoul_park.csv'))

order_week = ['월','화','수','목','금','토','일']


# 월별로 그룹화하여 평균값 계산
grouped_mon = df.groupby("요일")
# 각 월별로 수치형 컬럼들의 평균을 계산
df_week = grouped_mon.mean(numeric_only=True).reset_index()
# 요일 월화수목금 순서로 지정하기 위해 카테고리형으로 변환 후 정렬.
df_week['요일'] = pd.Categorical(df_week['요일'], categories=order_week, ordered=True)
df_week = df_week.sort_values('요일')
df_week
````

````python
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

fig, (cx1,cx2) = plt.subplots(1,2, constrained_layout=True)
# ---------------------
# (1) matplotlib 막대그래프
# --------------------
cx1.set_title("요일별 시각화 (matplotlib)")
cx1.bar(df_week['요일'], df_week['어른'], label='어른')
cx1.bar(df_week['요일'], df_week['어린이'], label='어린이')
cx1.bar(df_week['요일'], df_week['청소년'], label='청소년')
cx1.set_xlabel("요일")
cx1.set_ylabel("방문자 수")
cx1.legend()
cx1.grid(True, axis = 'y')

#--------------------
# (2) seaborn 막대그래프
#--------------------
cx2.set_title("요일별 시각화 (seaborn)")
# 데이터셋 불러올때 요일 순서 지정한대로 출력하기 위해 order 매개변수 전달
sns.barplot(data=df, x='요일', y ='어른',label = '어른',errorbar=None, ax = cx2, order=order_week)
sns.barplot(data=df, x='요일', y ='어린이',label = '어린이', errorbar=None, ax = cx2, order=order_week)
sns.barplot(data=df, x='요일', y ='청소년',label = '청소년', errorbar=None, ax = cx2, order=order_week)
cx2.set_ylabel("방문자 수")
cx2.grid(True, axis = 'y')
plt.show()
````

````python
'''
그룹으로 묶기: hue 매개변수 활용
- hue 매개변수에 그룹화할 컬럼 이름 전달
'''
sns.barplot(data=df, x='날씨', y ='어린이', errorbar=None, hue='공휴일')
plt.show()
````

## 다양한 그래프 그리기(3) : 파이 차트

파이차트
- 전체 데이터에서 그 범주에 해당하는 비율을 표현
    - 원형으로 표현되며, 부채꼴의 중심각은 해당 범주의 상대적인 비율
    - matplotlib의 pie 함수 사용(sseborn에는 별도로 없음)

## 다양한 그래프 그리기(4): 히스토그램

히스토그램
- 특정 구간에 해당하는 빈도를 표현하는 그래프
    - 이 구간에 속한 데이터가 몇개인지 표현
    - 특정 값의 빈도 분포를 분석할 때 주로 사용

````python
df = pd.read_csv('data/seoul_park.csv')
df_sat = df[df['요일'] == '토']
df_sat.describe()
````

````python
fig, (dx1, dx2) = plt.subplots(1,2, constrained_layout=True)

#----------
# matplotlib 히스토그램
#----------
dx1.hist(df_sat['어린이'], bins=10)
# bins : 구간의 개수(=막대의 개수) 설정
dx1.set_title("matplotlib 히스토그램")
dx1.set_xlabel("어린이")
dx1.set_ylabel("count")
dx1.grid(True, axis='y')

# ----------
# seaborn 히스토그램
# ----------
sns.histplot(data=df_sat, x='어른', bins=10, ax=dx2)
dx2.set_title("seaborn 히스토그램")
dx2.grid(True, axis='y')
plt.show()
````

````python
'''
히스토그램 : 그룹화

- hue 매개변수를 통해 그룹화
- 2월의 어린이 데이터를 연도별 분포를 보여주는 그래프
'''
df_feb = df[df['월'] == 2]
sns.histplot(data=df_feb, x='어린이', bins=10, hue='연')
plt.show()
````

````python
'''
히스토그램 : 구간의 범위 설정하기

- binrange 매개변수를 통해 전체 구간의 범위를 설정
- 지정하지x -> Default로 데이터의 최소, 최대 설정
'''

sns.histplot(data=df_feb, x='어린이', hue = '연',binrange=(0,1000), bins=10)
````

````python
'''
히스토그램 : 추가 옵션

- multiple 매개변수로 막대 겹침 방식 설정(dodge:겹치지 않게)
- shrink: 막대 사이의 간격 조절(0.8: 80% 크기)
'''

sns.histplot(
    data=df_feb, 
    x='어린이', 
    hue = '연', # 세부적으로 그룹화
    binrange=(0,1000), # 구간 범위 설정
    multiple='dodge', # 막대 겹침 방식 설정(dodge:겹치지 않게)
    shrink=0.8, # 기존의 80% 크기로 막대 너비 설정
    bins=10  # 구간의 개수(=막대의 개수)
    )
plt.xticks(range(0,1000,100))

# 2월 데이터에서 어린이 데이터의 구간마다의 빈도수를 연도별로 확인
plt.show()
````
