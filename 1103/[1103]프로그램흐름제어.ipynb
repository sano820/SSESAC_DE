## 리스트 자료형
여러 개의 값을 담을 수 있는 자료형
- 대괄호 쌍으로 표현하며, 각각의 원소마다 ,(콤마)로 구분한다.
- 각각의 원소에는 순사가 존재한다.

## 시퀀스 자료형
- 순서가 있으면서, 여러 원소들을 담고 있는 자료형

### 인덱스
- 시퀀스 자료형의ㅡ 각 원소의 위치(순서, 순번)를 인덱스라고 표현한다.
- 첫번째 위치는 0부터 시작한다.

````python
num_list = [1,2,3,4,5]
# index ==> 0,1,2,3,4
````

````python
str_var = "Hi! Python"
#index ==> 0123456789
````

원소 접근하기 - (1) 인덱싱: 값 읽기
- 인덱스를 바탕으로 해당 위치에 접근하는 기법
- ``` 문자열/리스트[인덱스 번호]```

````python
alpha = [1,3,5,7,9,11]
print(alpha[2])
````

````python
beta = "Hello Python!"
print(beta[4])
````

원소 접근하기 - (1) 인덱싱: 값 수정
- ```리스트[인덱스_번호] = 값```
- 문자열은 수정이 불가능! 문자열은 Immutable 타입이다

````python
num_list = [1,2,3,4,5]
num_list[1] = 20
num_list[4] = 50

print(num_list)
````

````python
word = "pyThoN"

word[2]='t'
word[5]='n'
print(word)
````

원소 접근하기 - (2) 슬라이싱(Slicing)
- 인덱스 범위를 통해 연속된 범위의 일부분을 가져오는 기법
- ```문자열/리스트[시작_인덱스:끝_인덱스]```
    - 끝 범위의 인덱스는 포함하지 않는다!

````python
num_list = [1,2,3,4,5]
print(num_list)
````

````python
str_var = "Hi! Python"

print(str_var[0:3])
````

원소 접근하기 - (3) 인데싱과 슬라이싱 확장
- 음수의 인덱싱, 싲가 범위 또는 끝 범위를 비울 수 있다!

````python
a = "Hello"
b = [3,7,10,15,20]

print(a[-1])   # 뒤에서 첫번째
print(b[:3])   # 처음부터 3번째 미만
print(b[1:])   # 1번째부터 마지막까지
````

````python
# Quiz! 아래 출력 결과는 어떤 결과가 나올까요?
a = "Hello World"
b = [4,5,3,2,1]

print(a[-2:])
print(b[1:-3])
````

시퀀스 자료형 활용 : in 연산자
- 시퀀스 자료형 안에 해당 원소가 포함되어 있는지 검사하는 연산자
- ``` 원소 in 시퀀스 자료형 ```

````python
sentence = "Hello World"
num_list = [1,3,5,7,9]

print("E" in sentence)
print(3 in num_list)
````

시퀀스 자료형 활용 : len() ㅎ마수
- 시퀀스 자료형의 길이를 구하는 함수(= 총 원소의 개수)
- ``` len(시퀀스 자료형) ```

````python
num_list =[1,2,3,4,5]
word = "Python"

print(len(num_list))
print(len(word))
````

## 리스트 자료형 활용
리스트 자료형의 다양한 기능 살펴보기

### 중첩리스트(Nested List)
- 리스트 내의 우너소가 리스트인 구조(=2차원 리스트)


````python
nested = [[1,2,3],[4,5,6],[7,8,9]]
print(len(nested))
print(len(nested[0]))
````

````python
print(nested[0][1])
print(nested[1][2])
````

리스트 자료형마다의 기능(메서드)를 살표보자!   
```리스트변수.메서드()```
- append(x)
- insert(i,x)
- remove(x)
- pop(i)
- sort()
- index(x)
- count(x)
- clear()

list.append(x)
- 현재 리스트에 원소x를 가장 마지막에 추가

````python
a = []
b = ['p', ' y', 'z']
a.append(20)
a.append(10)

b.append('h')
b.append('o')
b.append('n')

print(a)
print(b)
````

list.insert(i,x)
- 현재 리스트에 i번째 인덱스에 원소 x를 추가(삽입)

````python
a = [1,2,4,5]
a.insert(2,3)
print(a)
````

list.remove(x)
- 현재 리스트에서 처음 검색된 원소 x를 삭제
    - 앞에서부터 일치하는 값 검색 후, 처음 검색되는 값을 삭제

````python
num_list = [3,1,2,3]
num_list.remove(3)
print(num_list)
````

list.pop(i)
- 현재 리스트에서 i 번째 인덱스 원소를 삭제 후, 반환(=원소 빼내기)

- 인자에 아무 것도 넣지 않으면, default로 마지막 원소를 대상으로 수행

````python
num_list = [1,2,3,4,5]

print(num_list.pop(1))
print(num_list.pop(1))
print(num_list.pop(2))
````

````python
num_list = [1,2,3,4,5]
print(num_list.pop())
````

list.iindex(x)
- 현재 리스트에서 원소 x 해당하는 인덱스 검색
    - 앞에서 일치하는 값 검색 후, 처음 검색된 위치를 반환

````python
num_list = [3,1,4,1,2]

print(num_list.index(4))
````

주의할 점! - remove(), pop(), index()
- 검색/삭제 대상의 원소나 인덱스가 존재하지 않거나 범위를 초과하면  에러 발생!

````python
num_list = [1,2,3,4,5]

print(num_list.pop(7))
````

````python
num_list = [1,2,3,4,5]

print(num_list.remove(10))
````

````python
num_list = [1,2,3,4,5]

print(num_list.index(0))
````

list.count(x)
- 현재 리스트에서 원소 x의 개수를 반환

````python
num_list = [3,1,4,1,2,1,3]

print(num_list.count(1))
print(num_list.count(3))
````

list.sort()
- 현재 리시트의 원소들을 내부적으로 정렬
    - Defalut는 오름차순 정렬, reverse 매개변수에 true 전달 시, 내림차순 정렬

````python
num_list =[ 4,2,3,1,5]

num_list.sort()
print(num_list)

num_list.sort(reverse=True)
print(num_list)
````

list.clear
- 현재 리시트의 모든 원소들을 내부적으로 제거하는 기능(=빈 리스트로 만드는 기능)

````python
a = [5,4,2,1,6]
a.clear()

print(a)
````

## 조건문(if-elif-else)

만약에 조건이 참이면, '수행할 명령'들을 수행해라
```
1. 검사할 조건 명시. 조건의 결과가 True가 되는지 판별
2. if문 정의ㅡ의 마무리를 의미하는 콜론(:) 작성
3. 종속된 명령을 구분하기 위한 들여쓰기
4. 조건에 따른 분기 처리 명령
```

````python
var1 = 1

if var1 > 0:
    print(var1, "은 양수입니다.")

print("조건문 종료")
````

````python
var1 = -1

if var1 > 0:
    print(var1, "은 양수입니다.")

print("조건문 종료")
````

### 들여쓰기(Indentation)

- 파이썬에서는 공백도 일종의 문법이다.

- if/else/for/while 등등 블록의 들여쓰기는 공백 4칸 또는 탭 1칸을 의미한다!

````python
num1 = 10
num2 = 20

if num1 + num2 > 30:
    print(num1, num2)
````

````python
radius = int(input("반지름을 입력하세요: "))

if radius > 0:
    print('넓이 = ', (3.14 * radius**2))
    print('둘레 = ', (2 * 3.14 * radius))
````

````python
num = int(input('정수를 입력하세요: '))

if num % 2 ==0:
    print(num, '은 짝수입니다.')
````

### if-else 문
만약에 조건이 맞지 않을 때는? else 구문!
- 만약에 조건이 참이면 A 실행, 그렇지 않으면 B 실행

````python
'''
if 조건:
    print('A')
else:
    print('B')
'''
````

````python
num = int(input('정수를 입력하세요 : '))

if num%2 ==0:
    print(num, "은 짝수입니다.")
else:
    print(num, '은 홀수입니다.')
````

if-else 구문은 하나의 조건, 그리고 그렇지 않은 경우로 이분법적이다.
- 그렇지 않고, 그 이상의 여러 개의 조건을 걸고 싶을 때는? -> if-elif-else 문!

```
if 조건1:
    <if문 종속 명령>
elif 조건2:
    <elif문 종속 명령>
else:
    <else 종속 명령>
```

````python
grade = input("등급을 입력하세요.: ")


if grade == 'A':
    print('A등급입니다.')
elif grade == 'B':
    print('B등급입니다.')
elif grade == 'B':
    print('C등급입니다.')
elif grade == 'B':
    print('D등급입니다.')
else:
    print('F등급입니다.')
````

````python
# 조건을 세부적으로 걸기!
num1 = 10
num2 = -10

if num1 > 0:
    if num2 > 0:
        print('둘다 양수입니다.')
    elif num2 < 0:
        print('num1은 양수, num2는 음수입니다.')
elif num1 < 0:
    if num2 > 0:
        print('num1은 음수, num2는 양수입니다.')
    elif num2 < 0:
        print('둘다 음수입니다.')
````

비교 연산자를 한 줄에 이어서 쓸 수 있는 기능(조건 연결하기)  
``` if num >= 10 and num <= 90: ```

````python
a = ['H', 'h', 'e', 'l', 'l', 'o']

if 'H' in a:
    a.remove('H')
    print(a)
else:
    print("H가 없습니다")
````

````python
word = "Python"

if len(word) >= 5:
    print(word * 2)
else:
    print(word)
````

````python
word = input('단어를 입력하세요: ')

if len(word) >= 5 and 'P' in word:
    print(word)
else:
    print('조건에 맞지 않은 단어입니다.')
````

````python
score = int(input("시험 점수를 입력하세요.: "))
attendance = int(input("출석 점수를 입력하세요.: "))

if score >= 70 and attendance >= 90:
    print("통과")
elif score >= 70 and attendance < 90:
    print("조건부 통과(출석 미달)")
else:
    print("미흡")
````
