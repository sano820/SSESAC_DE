### 내장 함수(Built-in) 
= 파이썬 개발자들이 이미 만들어 둔 함수들(기본적으로 탑재가 되어있으므로 사용만 하면 됨)  

````python
'''
다양한 내장 함수(Built-in) - 집계 함수
max() = 시퀀스 자료의 최댓값을 구하는 함수
min() = 시퀀스 자료의 최솟값을 구하는 함수
sum() = 시퀀스 자료의 합을 구하는 함수
'''

# max()
print("max 함수 활용")
print(max((1,2,3,4,5)))  # 5
print(max([20,40,10,50,30]))  # 50

# min()
print("\nmin 함수 활용")
print(min((1,2,3,4,5)))  # 1
print(min([20,40,10,50,30]))  # 10

# sum()
print("\nsum 함수 활용")
print(sum((1,2,3,4,5)))  # 15
print(sum([20,40,10,50,30]))  # 150

````

### 사용자 정의 함수(User-defined)

사용자 정의 함수 만들기(1) : 정의하기

````python
def hello(x):  # def 키워드
    y = 10 * x + 2
    return y

hello(10)
````

### 함수 확장

Default Parameter(디폴트 매개변수)
- 매개변수에 디폴트 값을 명시하는 것으로, 매칭되는 인자 없이 호출 시 디폴트 값을 사용

````python
def def_para(country="Korea"):
    print('I am from', country)

def_para('India')
def_para('Brazil')
def_para()
````

````python
# 디폴트 매개변수는 매개변수들 중에서 마지막에!!

def def_score(name, kor, eng, mat=50):
    print(name, kor, eng, mat)

def_score('John', 90, 80, 70)
def_score('Rosa', 95, 85)
````

Variavle Argument(가변 매개변수)
- 여러 개의 값을 한번에 받을 수 있는 매개변수 (*args)
    - 매개변수 이름 앞에 *을 붙여주면 가변 매개변수가 된다.
    - 전달 받은 매개변수는 tuple 형태로 받은 값들을 묶는다.

````python
kor, eng, mat, sci = 98, 77, 85, 90

def max_score(*args):
    return max(args)

print(max_score(kor, eng, mat))
print(max_score(eng, mat, sci))
````

매개변수 이름으로 인자 전달(Keyword Argument)
- 함수 호출시, 각 인자에 매개변수의 이름을 명시해서 전달 가능
- **주의할점!** : 키워드 인자가 나오면 그 뒤의 모든 인자는 반드시 키워드 형태!!

````python
def sum_score(kor,eng, math):
    return kor + eng + math

print(sum_score(kor=70, eng = 80, math = 90))
````

함수와 Unpacking/Packing 활용
- 함수의 인자로 리스트에 *을 붙여주면 unpacking을 수행
- 함수의 반환 부분에 여러 개의 값을 반환 가능

````python
def unpacking(a,b,c):
    print(a)
    print(b)
    print(c)


mylist = [10,20,30]
unpacking(*mylist)
````

````python
def three(x):
    return x, x**2 , x**3

a,b,c = three(3)
print(a,b,c,)
````

## 지역변수와 전역변수

전역변수 == 전역적인 범위에서 사용 가능한 변수

````python
global_var = "Hi"

def my_func(a):
    print(a)
    print(global_var)

my_func(global_var)
print(global_var)
````

지역변수 == 특정 구문(for문, 함수) 안에서 정의한 변수
- 변수를 정의한 범위에서만 사용 가능

````python
def my_func():
    local_var = "Hello"
    print(local_var)

my_func()
print(local_var)
````

### 프로세스 메모리 구조(Python)
- Code = 프로그램의 명령
    - bytecode가 저장되는 영역
- Heap = 모든 객체(실제 값)들이 저장되는 영역
- Stack = 지역변수 배개변수, 함수 호출 정보가 저자되는 영역
- *Global namespase* = 전역변수의 {이름 : 값} 쌍이 저장되는 영역 (논리적 영역)


전역 변수는 Global Namespace 영역에 디셔너리 형태로 저장.
- stack 영역에 Global Namesapce에 대한 포인터(참조)가 있어서 함수 내부에서 전역 변수 참조

````python
global_var = 10

def my_func():
    print(global_var)
    global_var += 10
    print(global_var)

my_func()
````

````python
global_var = 10

def my_func():
    global global_var
    print(global_var)
    global_var += 10
    print(global_var)

my_func()
````


## 클래스(Class)와 객체(Object)

### 객체 지향 프로그래밍(Object-Oriented Programming)
- 객체를 중심으로 프로그래밍하는 패러다임.
- 모든 것을 객체(object)로 간주.

Class = 객체들의 유사한 특징을 지닌 속성과 동작을 묶어놓는 집합

Example : "사람"  
- Q. 사람의 공통 속성?
- Q. 사람의 공통 동작?

클래스 == 설계도   <-->  객체 == 설계도를 바탕으로 만들어진 실제 상품

- 프로그래밍상 실제 메모리상에서 생성된 객체를 인스턴스라고 함.

````python
'''
클래스 예시 1 : 붕어빵 클래스

속성 : 속 재료, 속의 양
동작 : 속 채우기, 붕어빵 꺼내기, 속 양 조절하기
'''
class Bungeoppang:
    def __init__(self, filling="팥", filling_amount=50):
        """
        붕어빵 초기화
        filling: 속재료 (예: 팥, 슈크림, 피자 등)
        filling_amount: 속의 양 (0~100 기준)
        """
        self.filling = filling
        self.filling_amount = filling_amount
        self.is_filled = False
        self.is_taken_out = False

    def fill_filling(self):
        """속을 채우는 동작"""
        if self.is_filled:
            print("이미 속이 채워져 있습니다!")
        else:
            self.is_filled = True
            print(f"{self.filling} 속을 {self.filling_amount}% 만큼 채웠습니다!")

    def adjust_filling_amount(self, amount):
        """속의 양 조절"""
        if 0 <= amount <= 100:
            self.filling_amount = amount
            print(f"속의 양을 {self.filling_amount}%로 조절했습니다.")
        else:
            print("속의 양은 0~100 사이여야 합니다!")

    def take_out(self):
        """붕어빵 꺼내기"""
        if not self.is_filled:
            print("속을 아직 채우지 않아서 꺼낼 수 없습니다!")
        elif self.is_taken_out:
            print("이미 꺼낸 붕어빵입니다!")
        else:
            self.is_taken_out = True
            print(f"{self.filling} 붕어빵을 꺼냈습니다. 맛있게 드세요!")

    def info(self):
        """붕어빵 상태 출력"""
        status = "속 채움 완료" if self.is_filled else "속 미채움"
        taken = "꺼냄" if self.is_taken_out else "안 꺼냄"
        print(f"[붕어빵 정보] 속재료: {self.filling}, 속 양: {self.filling_amount}%, 상태: {status}, 꺼내기 여부: {taken}")

````

````python
bbang = Bungeoppang(filling="슈크림", filling_amount=70)

bbang.info()
bbang.fill_filling()
bbang.adjust_filling_amount(90)
bbang.take_out()
bbang.info()

````

````python
class Car:
    def __init__(self, color="흰색", fuel_efficiency=15.0, model="기본형"):
        self.color = color
        self.fuel_efficiency = fuel_efficiency
        self.model = model
        self.is_engine_on = False
        self.speed = 0
        self.gear = "P"  # 기본 기어 상태 (P: 주차)

    def start_engine(self):
        """시동 걸기"""
        if self.is_engine_on:
            print("이미 시동이 켜져 있습니다.")
        else:
            self.is_engine_on = True
            print(f"{self.model}의 시동이 켜졌습니다.")

    def accelerate(self, amount):
        """가속하기"""
        if not self.is_engine_on:
            print("시동이 꺼져 있어서 가속할 수 없습니다.")
        elif self.gear not in ["D", "R"]:
            print("주행 기어(D 또는 R)가 아닙니다. 기어를 변경하세요.")
        else:
            self.speed += amount
            print(f"{self.model}이(가) {amount}km/h 가속하여 현재 속도는 {self.speed}km/h 입니다.")

    def stop(self):
        """멈추기"""
        if self.speed > 0:
            self.speed = 0
            print(f"{self.model}이(가) 멈췄습니다.")
        else:
            print("이미 정지 상태입니다.")

    def change_gear(self, gear):
        """기어 변경하기"""
        valid_gears = ["P", "R", "N", "D"]
        if gear in valid_gears:
            self.gear = gear
            print(f"기어를 {self.gear}(으)로 변경했습니다.")
        else:
            print("잘못된 기어입니다. (가능한 값: P, R, N, D)")

    def info(self):
        """자동차 상태 출력"""
        engine_status = "ON" if self.is_engine_on else "OFF"
        print(f"[자동차 정보]")
        print(f"모델명: {self.model}")
        print(f"색상: {self.color}")
        print(f"연비: {self.fuel_efficiency} km/L")
        print(f"시동: {engine_status}")
        print(f"속도: {self.speed} km/h")
        print(f"기어: {self.gear}")
````

````python
car1 = Car(color="빨간색", fuel_efficiency=12.5, model="소나타")

car1.info()
car1.start_engine()
car1.change_gear("D")
car1.accelerate(30)
car1.stop()

car1.info()

````

### Python에서의 클래스(Class)

````python
'''
클래스 정의 및 객체 생성하기
'''

class Person:                           # 클래스 이름
    def __init__(self, name, age):      # 클래스의 동작(=메서드)
        self.name = name                # 클래스의 속성1
        self.age = age                  # 클래스의 속성2

p1 = Person("John", 36)                 # 객체 생성

print(p1.name, p1.age)
````

Class 정의하기  
1. Class 이름
    - Class 이름은 PascalCase로 짓는 것이 관례적
    - ex) Car, SmartPhone
    - 생성자(Construnctor)
        - 객체(인스턴스)가 생성될 때, 한번 호출되는 함수
        - 이름은 반드시 "__init__' 형태
        - 역할 : Class의 속성을 정의하고ㅡ, 초기화하는 작업에 주로 사용
2. 인스턴스 변수
    - 클래스의 속성으로 정의된 변수를 의미하며 객체마다 서로 다르다
    - 객체 생성 후, 인스턴스 멤버 변수에 접근하는 방법은 ```객체.인스턴스변수```형태
    - self 키워드
        - 객체 자기 자신을 의미하는 키워드
        - 실제 객체가 생성되었을 때, self가 해당 객체 자기자신이 되는 것
3. 인스턴스 메서드
    - 클래스 내부에 정의된 함수를 메서드
        - 특정 주체가 해당 동작을 수행. ```객체.메서드()``` 형태
        - !! 모든 인스턴스 메서드에는 self 매개변수가 포함되어야 한다!

````python
'''
사람 클래스 정의(선언부)
- name과 age 멤버 변수(속성)
- walk와 eat 멤버 메서드(동작)
'''
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def walk(self):
        print(f"{self.name}이 걷습니다.")
    
    def eat(self, food):
        print(f"{self.name}이 {food}를 먹습니다.")

# 객체 생성
gildon = Person("Gildon", 22)
john = Person("Jogn", 40)

# 객체 활용(변수, 메서드 호출 등)
gildon.walk()
john.eat("Banana")
````

````python
'''
Car 클래스를 정의하고, 생상자로부터 두가지 속성의 정의하고 두가지의 메서드를 정의,
그리고 두 객체를 생성 후 , 오른쪽 아래와 같이 출력하도록 코드를 작성하시오.

- 클래스 이름 : Car
- 속성 : name, speed
- 메서드 : get_name, get_speed
    - get_name : name 속성 값을 반환
    - get_speed : speed 속성 값을 반환
'''

class car:
    def __init__(self, name = "KIA", speed = "10"):
        self.name = name
        self.speed = speed
        
    def get_name(self):
        print(f"Car name : {self.name}")

    def get_speed(self):
        print(f"Speed : {self.speed}")

    def info(self):
        self.get_name()
        self.get_speed()
        print(f"{self.name}'s speed is {self.speed}")
        print("=======================")

a = car("Audi", 20)
b = car("BMW", 30)
c = car()

a.info()
b.info()
c.info()
````

````python
# Solution
class Car:
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed
    def get_name(self):
        return self.name
    def get_speed(self):
        return self.speed
    
car1 = Car("Audi", 10)
car2 = Car("BMW", 30)

print(f"{car1.get_name()}'s speed is {car1.get_speed()}")
print(f"{car2.get_name()}'s speed is {car2.get_speed()}")
````
