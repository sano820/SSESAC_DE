## 다양한 그래프 그리기 (5) : 박스 플롯

### 박스 플롯의 이해

박스 플롯
- 데이터의  분포를 보여주는 식가화 그래프
    - 수치형 변수의 데이터를 요약하고 비교하는데 사용
    - 단순하게 평균을 보면 알 수 없는 데이터의 중앙값, 사분위수, 이상치 등의 분포를 확인할 수 있음

박스 플롯의 구성요소(1) : 박스(BOX)
- 데이터의 중앙값과 사분위수를 나타냄
    - 가운데 선이 중앙값
    - 박스의 상단과 하단을 각각 Q1(상위 25%), Q3(하위 25 %)를 나타냄
    - 박사의 높이는 IQR(interquartile Range)로, Q3-Q1
        - 즉, 데이터의 50% 부분을 의미
  
박스 플롯의 구성요소(2) : 수염(whisker)
- 박스의 범위를 벗어난 데이터
    - Q3 + IQR * 1.5 // Q1 - IQR * 1.5

박스 플롯의 구성요소(3) : 이상치(Outlier)
- 수염의 범위를 벗어난 값
- 잘못된 축정으로 인해 발생한 동 떨어진 극단적인 값
- 보통 원 마름모 등등의 도형으로 표현

### 그래프 그려보기

```python
plt.boxplot(df_2018_apr['어린이'])
# 어린이 컬럼을 선택하여 이 데이터에 대한 박스 플롯 그림
```

### 참고: 데이터 프레임에서 이상치 추출하기

```python
Q1 = df['value'].quantile(0.25) # 25% 분위수
Q3 = df['value'].quantile(0.75) # 75% 분위수
IQR = Q3-Q1

outlier = df[(df['value'] < Q1 - 1.5 * IQR) | (df['value'] > Q3 + 1.5 * IQR)]
# IQR(Interquartile Range) 기법을 사용한 이상치 탐지
```

## 다양한 그래프 그리기 (7) : 히트맵

히트맵
- 데이터 값의 상대적인 크기를 색상으로 표시하여 시각화하는 그래프
    - 2차원 표 형태의 데이터를 사용하여 생성
    - 값의 크기에 따라 다른 색상 맵을 사용하며, 값이 클수록 진한 색, 작을수록 연한 색 등으로 표현 됨.

## 탐색적 데이터 분석(EDA)

본격적인 모델링에 앞서, 데이터의 구조, 패턴, 관계, 이상치, 결측치, 분포 등을 탐색해 데이터의 상태/특징/관계/문제점을 파악하는 단계

EDA 과정에서 파악할 요소
- 기본적인 데이터를 이해하는 과정!
    - 어떤 데이터인가? 구조가 어떠한가?
    - 데이터에 오류(결측치, 이상치 등)는 없는가?
    - 각 변수의 기술적인 통계량이 어떠한가?
    - ...등등

### 위치 추정

각 변수를 대표하는 값(Central Tendency)을 구하여 데이터를 살펴보는 것

|유형|설명|특징|
|---|---|---|
|평균|모든 값을 더해 개수를 나눈 값|Outlier에 민감|
|가중 평균|각 값의 중요도를 고려한 평균|ex. 시험 점수 평균 <br> (중간40%, 기발 60%)|
|중앙값|크기순으로 정렬했을 때, 중간의 값|outlier의 영향이 적음|
|최빈값|가장 자주 등장한 값|범주형 데이터|

### 변이 추정


데이터가 얼마나 퍼져 있는지 살펴보는 것(Dispersion)

|유형|설명|특징|
|---|---|---|
|분산|평균으로부터 얼마나 퍼져 있는지 제곱 단위로 나타낸 값|-|
|표준편차|분산의 제곱근으로 데이터의 흩어짐 정도를 원래 단위로 표현한 값|-|
|수반할 볌위|75번째 백분위수와 25번째 백분위수의 창이|중앙 50% 구간의 폭|

### 6. 다변량 분석(여러 개의 변수 간의 관계 분석)
- 타겟 변수인 Survived와 다른 변수와 관계를 파악

```python
# 6-1. 생존여부와 pclass와의 관계
sns.countplot(data = df, x='pclass', hue = 'alive')
```

```python
fig, (dx1, dx2) = plt.subplots(1,2)

# pclass와 생존 여부에 대한 비율(pclass와 사망에 대한 비율)
# 사망자수는 survived의 컬럼이 0인 행
dead_pclass = df[df['survived']==0]['pclass'].value_counts().reset_index()
print(dead_pclass)
dx1.pie(x = dead_pclass['count'], labels = dead_pclass['pclass'], autopct="%.1f%%")
dx1.set_title("사망자의 선실등급 비율")

# 생손자의 선실등급 비율
live_pclass = df[df['survived']!=0]['pclass'].value_counts().reset_index()
dx2.pie(x=live_pclass['count'], labels = live_pclass['pclass'], autopct = "%.1f%%")
dx2.set_title("생손자의 선실등급 비율")
plt.show()
```

```python
# 6-2 성별에 따른 생존 여부 탐색
# 성별에 따른 평균 생존률
print(df.groupby('sex')['survived'].mean())

# 분포 확인
sns.countplot(data = df, x = 'sex', hue = 'survived')
```

```python
# [Todo] 성별에 따른 평균 생존률을 파이 차트로!
# 여성따로, 남성 따로!
mean_sex = df.groupby('sex')['survived'].mean().reset_index()
print(mean_sex)

plt.pie(x=mean_sex['survived'], labels = mean_sex['sex'], autopct = '%.1f%%')
plt.show()
```

```python
#6-3 나이에 따른 생존 여부 탐색
fig, (ex1,ex2) = plt.subplots(1,2, constrained_layout=True)

sns.histplot(data = df, x = 'age', hue = 'survived', ax=ex1)
ex1.set_title('연령대별 생존률')



# 구간마다의 생존률
tmp = df.dropna(subset = ['age', 'survived']).copy()
# 구간의 범위를 지정
bins = [0,10,20,30,40,50,60,70,100]
labels = ['0-9', '10-19','20-29','30-39','40-49','50-59','60-69','70+']
tmp['age_r'] = pd.cut(tmp['age'], bins=bins, labels=labels)
#cut 함수 : 연속형 데이터를 구간(범주)으로 나눈 함수
age_rate = tmp.groupby('age_r')['survived'].mean().mul(100).reset_index()

sns.barplot(data = age_rate, x = 'age_r', y = 'survived', ax = ex2)
ex2.set_xticklabels(ex2.get_xticklabels(), rotation=45)
ex2.set_ylim(0,100)
ex2.set_title('나이에 따른 생존률')
plt.show()
```

```python
# 6-4. 타켓 변수와 sibps, parch (형제자매와 자녀, 부모의 수)// 즉 관계의 수
fig, (fx1,fx2) = plt.subplots(1,2, constrained_layout=True)

print(df['sibsp'].value_counts())
sns.countplot(data=df, x = 'sibsp', hue = 'survived', ax =fx1)
fx1.set_title('함께 탐승한 형제/자메, 배우자의 총합')


print(df['parch'].value_counts())
sns.countplot(data=df, x='parch', hue='survived', ax = fx2)
fx2.set_title('함께 탑승합 부모, 자녀의 총합')


plt.show()
```

```python
# sibsp + parch
df['family'] = df['sibsp'] + df['parch']

# 가족의 수에 따른 평균 생존률 시각화
family_df = df.groupby('family')['survived'].mean().reset_index()
print(family_df)
sns.barplot(data = family_df, x = 'family', y = 'survived')
plt.ylim(0,1)
```

```python
# 6-5. 타겟변수와 Embarked 탑승구 위치에 대한 탐색
fig, (gx1,gx2) = plt.subplots(1,2, constrained_layout=True)

sns.countplot(data = df, x = 'embarked', hue = 'survived', ax = gx1)

# pclass와 embarked에 따른 생존 여부
# 왜 s 항구에서 사망이 제일 많았을까? -> 티켓 등급과 연결해서 확인
freq = pd.crosstab(df['pclass'], df['embarked'])
# crosstab : 두 개 이상의 범주형 변수 간의 교차 빈도표를 생성
print(freq)
# 원래는 tnstjrk C-Q-S인데, S-Q-C로 바꾸는 작업
freq = freq[['S','C','Q']]

freq.plot(kind = 'bar', stacked=True, ax=gx2) # 막대그래프인데 누적 막대 그래프

plt.show()
```
