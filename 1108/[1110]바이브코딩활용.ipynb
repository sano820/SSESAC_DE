### pack() 메서드
- 레이아웃 관리자로 위젯을 위에서 아래로 순서대로 배치
    - 위젯에 대한 pack을 호출하지 않으면 위젯이 실제로 보이지 않음

- side = LEFT/RIGHT/TOP/BOTTM (위젯을 어느 방향에 붙일 것인지)


- fill = NONE/X/Y/BOTH ( 공간을 채우는 방향),
- expand = 남는 공간 있으면 여유 공간을 모두 채울지 여부


### 다양한 위젯

Label - 텍스트나 이미지 등을 화면에 출력하기 위한 위젯


````python
from tkinter import *

win = Tk()

lbl1 = Label(win, bg = 'green', fg='white')
lbl1['text'] = 'tkinter'
lbl1.pack()

lbl2 = Label(win, text='Python', bg='red', font=('Helvetica',166))
lbl2.pack()

win.mainloop()
````

Button - 사용자가 클릭했을 때, 특정 동작을 실행하는 인터렉티브 위젯

````python
from tkinter import *
win = Tk()

def click():
    print('Button Clicked')

btn = Button(win, text = 'click', bg = 'red', fg = 'white', command = click)
btn.pack()

win.mainloop()
````

````python
# .get 메서드를 통해 입력한 문자열을 가져올 수 있다.

from tkinter import *

win = Tk()

def click():
    print(entry.get())

entry = Entry(win)
entry.pack(fill = X)

btn = Button(win, text = 'view', command = click)
btn.pack(fill=X)

win.mainloop()
````

````python
# 입력하는 거를 *로 가리기
from tkinter import *

win = Tk()

def click():
    print(entry.get())

entry = Entry(win, show = '*')
entry.pack(fill = X)

btn = Button(win, text = 'view', command = click)
btn.pack(fill=X)

win.mainloop()
````

checkbutton 
- 사용자가 클릭할때마다 선택(on/off) 상태가 바뀌는 버튼
    - 내부적으로는 True/False(또는 1/0) 값을 가진다.
    - ```chk = Checkbutton(win, 옵션, ...)```
        - command : 체크 상태가 바뀔 때 실행할 함수
        - text : 체크박스 옆에 표시되는 텍스트
        - variable : 체크 상태를 저장할 변수(IntVar, BooleanVar, StringVar)

    - select(): 체크박스의 체크 상태(on)로 설정
    - deselect() : 체크박스의 체크 해제(off)로 설정

````python
from tkinter import *

win = Tk()

def check():
    print(var.get())
var = IntVar()

chk = Checkbutton(win, text = ' agreement', variable = var, command = check)
chk.pack()

win.mainloop()
````

````python
from tkinter import *

win = Tk()

def check():
    print(var1.get(), var2.get())

var1 = IntVar()
var2 = IntVar()

chk1 = Checkbutton(win, text = 'Music', variable = var1, command=check)
chk1.select()
chk1.pack(side=LEFT)

chk2 = Checkbutton(win, text='Video', variable=var2, command=check)
chk2.deselect()
chk2.pack(side=LEFT)

win.mainloop()
````

Radiobutton
- 라디오 버튼의 그룹, 한가지 항목만 선택 가능
- ```radio = Radiobutton(win, 옵션, ...)```
    - text : 체크박스 옆에 표시되는 텍스트
    - variable : 체크 상태를 저장할 변수
    - value : 라디오 버튼의 값

````python
from tkinter import *

win = Tk()

def check():
    print(g1.get())

g1 = StringVar()

rd1 = Radiobutton(win, text = 'Male', variable=g1, value = "M", command=check)
rd1.pack(side=LEFT)

rd2 = Radiobutton(win, text = 'Female', variable=g1, value = "F", command=check)
rd2.pack(side=LEFT)

rd1.select()

win.mainloop()
````

Listbox
- 여러 개의 항목을 나열하고, 그 중 하나(또는 여러 개)를 선택 가능한 위젯
- ```lst = Listbox(win, 옵션, ...)```
    - height : 표시할 최대 줄 수
    - selectmode : 선택 모드
- curselection() : 현재 선택된 인덱스 반환
- insert(index,item) : 항목 추가
- get(index) : 특정 항목 가져오기(순서와 텍스트가 담긴 하나의 투플)

````python
'''
리스트 중 하나만 선택
'''
from tkinter import *

win = Tk()

def select():
    index = lst.curselection()[0]
    print(index, lst.get(index))

lst = Listbox(win, height =3, selectmode = SINGLE)
lst.insert(0, 'red')
lst.insert(1, 'green')
lst.insert(2, 'blue')
lst.pack()

btn = Button(win, text='click', command = select)
btn.pack()

win.mainloop()
````

````python
'''
리스트 중 하나(또는 여러 개)를 선택
'''

from tkinter import *

win = Tk()

def select():
    for index in lst.curselection():
        print(index, lst.get(index))

lst = Listbox(win, height = 3, selectmode = MULTIPLE)
color = ['red','green','blue']

for item in color:
    lst.insert(END, item)
lst.pack()

btn = Button(win, text='click', command=select)
btn.pack()

win.mainloop()
````
