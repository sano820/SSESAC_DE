## 자료형의 연산

숫자형 자료형의 기본 연산자
 - \+   : 더하기 
 - \-   : 빼기  
 - \*   : 곱하기 
 - \/   : 나누기

```python
print(10+10)  # 20
print(10-5)  # 5
print(3*5)  # 15
print(9/4)  # 2.25
```

숫자형 자료형의 **특수**연산자  
- //  : 몫 나눗셈 연산자
- %   : 나머지 연산자
- **  : 제곱 연산자

```python
print(20 // 4) # 5
print(9 // 4) # 2
print(17 // 4) # 1
print(2 // 4) # 16
```

연산자의 특징 - 순환하거나 주기적인 패턴이 띄는 곳에서 주로 활용
ex) 시계 계산, 요일 계산, 리스트 인덱스 순환, 배수

```python
hour = 23
after = 5
new_hour = (hour + after) % 24
print(new_hour)
```

(3) 연산자의 확장
- 연산 수식 축약 표현 : 변수 자기 자신에게 연산을 하는 경우

|종류|기존 수식|축약 수식|
|---|---|---|
|덧셈|a = a + 1| a += 1|
|뺄셈|a = a - 1| a -= 1|
|곱셈|a = a * 1| a *= 1|
|나눗셈|a = a / 1| a /= 1|

*다른 연사자도 가능합니다.*

```python
total1 = 0
total2 = 0

total1 = total1 + 5
total2 += 5

print(total1)
print(total2)
```

```python
a = 1
b = 1

a = a * 2
b *= 2

print(a)
print(b)
```

# 논리 자료형과 비교/논리 연산자


## 논리 자료형(Boolean) 
- 참(True) 또는 거짓(False) 값을 갖는 자료형(bool 타입)  
*True는 1, False는 0으로도 간주된다.

```python
print(True)
print(False)
print(type(True))
```

다양한 값들에 대한 명제 안에서의 비교를 통해 논리 자료형의 값이 결과로 나옴

```python
print(3 < 5) # 주어진 명제가 참이면 True
print(7 == 5) # 주어진 명제가 거짓이면 False
print(1 >= 5) # False
```

## 비교 연산자
- 명제의 두 값을 비교하여 결과를 논리 자료형 형태로 반환

| 연산자 | 설명 | 예시 | 결과 |
| --- | --- | --- | --- |
| == | 같다 | 3 == 3 | True |
| ≠ | 같지 않다 | 3 ≠ 5 | True |
| > | 왼쪽이 더 크다 | 3>5 | False |
| < | 오른쪽이 더 크다 | 3<5 | True |
| ≥ | 왼쪽이 크거나 같다 | 3≥5 | False |
| ≤ | 오른쪽이 크거나 같다 | 3≤5 | False |

- 값 비교를 반드시 숫자형만 가능한 것이 아니다!
- 문자열, 그리고 그 뿐만 아니라 다른 자료형도 가능하다.

### 아스키코드(ASCII Code)
컴퓨터가 문자(알파벳, 숫자, 기호)를 숫자로 약속한 표

```python
'''
문자 A = 65
문자 B = 66

문자 a = 97
문자 b = 98
'''
print("A" < "B")
print("a" < "b")
print("A" < "a")
print("aaa" < "aab")  
print("aab" < "aba")  # 9798 < 9897
```

### 논리 연산자(Logical Operators)
여러 조건을 동시에 확인해야할 때, 복잡한 조건을 표현할 때 사용되는 연산자

1. and 연산자
    - 두(또는 그이상) 명제가 **모두 참**이어야 참이 되는 연산자

|경우|결과|
|---|---|
|True and True|True|
|True and False|False|
|False and True|False|
|False and False|False|

```python
'''
3==3 : 3은 3와 같은가요?        True
4<=5 : 4는 5와 작거나 같은가요?  True
6>2  : 6은 2보다 큰가요?        True
'''
print(3==3 and 4<=5 and 6>2)
```

2. or 연산자
    - 두(또는 그이상) 명제 중 **하나라도 참**이면 참이 되는 연산자

|경우|결과|
|---|---|
|True and True|True|
|True and False|True|
|False and True|True|
|False and False|False|

```python
'''
3==4 : 3은 4와 같은가요?        False
4<=5 : 4는 5와 작거나 같은가요?  True
6<2  : 6은 2보다 작은가요?      False
'''
print(3==4 or 4<=5 or 6<2)
```

3. not 연산자
- 명제의 결과를 뒤집음

```python
'''
3은 4와 같은 것   =>  False
3은 4와 같은 것이 아니다 => Ture
'''
print(3==4)
print(not 3==4)
```

4. 연산의 우선순위

|우선순위|연산자|
|---|---|
|1|()|
|2|**|
|3|*, /, %, //|
|4|+, -|
|5|>, >=, <, <=|
|6|==, !=|
|7|and, or, not|

```python
'''
1. 괄호에 의해 3+4
2. 곱셉에 의해 7*2
3. >에 의해 14>10
'''

result = (3 + 4) * 2 > 10
print(result)
```

```python
'''
1. 첫번째 괄호의 a*b
2. 첫번째 괄호의 10>8
3. 두번째 괄호의 c//b
4. 두번째 괄호의 5==5
5. and 연산의 True and True
'''

a=5
b=2
c=10
result = (a*b>8) and (c // b == 5)
print(result)
```
